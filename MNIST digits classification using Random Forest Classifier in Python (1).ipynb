```python
#visualize the data
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
label = labels.loc[0]
pixel = pixels.loc[0]
pixel = np.array(pixel, dtype='uint8')
pixel = pixel.reshape((28,28))
plt.title('Label is{label}'.format(label=label))
plt.imshow(pixel, cmap='gray')
plt.show()
```

```python
#real code start from here
#spliting the data into training and testing images
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(mnist.data, mnist.target, test_size=1/7.0) #size is used for 10k testing data
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
#check accuracy score with x and y tests
model.score(x_test, y_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred, y_test)
```
